# Introduction

Welcome to the Dask Tutorial.

Dask is a parallel computing library that scales the existing Python ecosystem. This tutorial will introduce Dask and parallel data analysis more generally.

Dask can scale down to your laptop laptop, and up to a cluster. Accordingly, the tutorial comes in two pieces. In the first part, we'll use the environment you setup to analyze medium sized datasets in parallel on your laptop.

For the second half, you'll log into a jupyterhub deployment that will provide you with your own Dask cluster to solve even larger problems.
